```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
temperature=np.array([22.56,23.43,24.3,25.5,26.625])
horizontal_width=np.array([146,136,113,98,82])
vertical_width=np.array([188,171,155,139,113])
```

```python
plt.plot(temperature,horizontal_width,"o")
plt.plot(temperature,vertical_width,"o")
lh=linregress(temperature,horizontal_width)
lv=linregress(temperature,vertical_width)
plt.plot(temperature,lh.slope*temperature+lh.intercept)
plt.plot(temperature,lv.slope*temperature+lv.intercept)
```
